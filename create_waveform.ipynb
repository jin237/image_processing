```python
import numpy as np
from matplotlib import pyplot as plt
import soundfile as sf
 
tmfile = '/Users/jin237/Documents/juxtaphony_project/soundfile/sasayaki_throat.wav'
cmfile = '/Users/jin237/Documents/juxtaphony_project/soundfile/sasayaki_conde.wav'
data_tm, samplerate_tm = sf.read(tmfile)
data_cm, samplerate_cm = sf.read(cmfile)
t_tm = np.arange(0, len(data_tm))/samplerate_tm
t_cm = np.arange(0, len(data_cm))/samplerate_cm
plt.figure(dpi=150)
fig = plt.gcf()
size = fig.get_size_inches()*fig.dpi
print(size)
plt.xlabel('Time[s]')
plt.ylabel('Sound Level')
plt.margins(x=0,y=0.1)
plt.plot(t_tm, data_tm, label="Sasayaki", color='blue' ,alpha=0.6)
plt.plot(t_cm, 0.9*data_cm, label="Meikaku", color='red' ,alpha=0.6)
plt.grid()
plt.legend()
plt.savefig('mix_sasayaki.eps')
```
