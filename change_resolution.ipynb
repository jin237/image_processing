```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 20)
m = 1.5
c = 2
y = m*x + c
```

```python
plt.figure(dpi=150)
plt.plot(x, y)
plt.title("y=mx+c")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
fig = plt.gcf()
size = fig.get_size_inches()*fig.dpi
print(size)

plt.show()
```
